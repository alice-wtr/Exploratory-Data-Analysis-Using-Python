```python
def lesser_of_two_evens(a, b):
  if a%2 == 0 and b%2 == 0:
    return min(a,b)
  else:
    return max(a,b)



def animal_crackers(a):
  for i in range(len(a)):
    if a[i] == " ":
      b = i+1
      break

  if a[0].upper() == a[b].upper():
    return True
  else:
    return False



def makes_twenty(a, b):
  if a == 20 or b == 20:
    return True
  elif (a+b) == 20:
    return True
  else:
    return False
```

```python
def spy_game(nums):
  l = len(nums)

  first_zero = l
  second_zero = l

  for i in range(l):

    if nums[i] == 0 and first_zero < l and second_zero == l:
      second_zero = i

    if nums[i] == 0 and first_zero == l and second_zero == l:
      first_zero = i


  last_seven = -1
  for i in range(l):

    if nums[l-i-1] == 7:
      last_seven = l-i-1
      break

  if second_zero < last_seven:
    return True

  return False






#def main():

#  print(spy_game([1, 2, 3, 4, 5]))
#  print(spy_game([1, 0, 0, 7]))
#  print(spy_game([4, 5, 6, 7]))
#  print(spy_game([0, 0, 1, 7]))
#  print(spy_game([0, 7]))
#  print(spy_game([1,2,4,0,0,7,5]))
#  print(spy_game([1,0,2,4,0,5,7]))
#  print(spy_game([1,7,2,0,4,5,0]))

# main()
```

```python
def isprime(a):

  if a == 0 or a == 1:
    return False

  for i in range((int)(a/2)+ 1):

    if i == 0 or i == 1:
      continue

    if a%i == 0:
      return False

  return True

def count_primes(a):

  k = 0

  for i in range(a+1):

    if isprime(i):
      k = k+1

  return k

#def main():

#  print(f" 100: {count_primes(100)}")
#  print(f" 12: {count_primes(12)}")
#  print(f" 1: {count_primes(1)}")
#  print(f" 0: {count_primes(0)}")
#  print(f" 2: {count_primes(2)}")
#  print(f" 10: {count_primes(10)}")

#main()
```
